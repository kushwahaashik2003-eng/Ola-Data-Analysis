```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert Date and Time
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time  # correct format

# Bookings per day
daily_bookings = df.groupby(df['Date'].dt.date)['Booking_ID'].count().reset_index()
daily_bookings.columns = ['Date', 'Total_Bookings']

# Seaborn lineplot
plt.figure(figsize=(15,5))
sns.lineplot(data=daily_bookings, x='Date', y='Total_Bookings', marker='o', color='red')

# Chart formatting
plt.title("Daily Bookings Trend", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Number of Bookings", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


```

```python
avg_data= {
    'Category':['Customer Rating', 'Driver Rating'],
    'Average':[
            df['Customer_Rating'].mean(),
            df['Driver_Ratings'].mean()
    ]
}

plt.figure(figsize=(7,6))
ax=sns.barplot(data=avg_data, x='Category', y='Average',hue='Category', palette='bright')
for container in ax.containers:
      ax.bar_label(container, label_type='edge', padding=2)
    
plt.title("AVG Customer VS Driver Rating",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example data
avg_data = {
    'Category': ['Customer Rating', 'Driver Rating'],
    'Average':[ df['Customer_Rating'].mean(),
              df['Driver_Ratings'].mean()]
}

# Colors and explode settings
colors = ['#66b3ff', '#ffcc99']
explode = (0.01, 0.01)  #     

plt.figure(figsize=(7,7))
plt.pie(avg_data['Average'],
        labels=avg_data['Category'],
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        shadow=True,           # Shadow effect
        explode=explode,       # Slight 3D separation
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.2})

plt.title("AVG Customer VS Driver Rating", fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
